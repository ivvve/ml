#### neighbourhood_group
- location

```python
"""
group
- Manhattan
- Brooklyn
- Queens
- Bronx
- Staten Island

(Manhattan, Brooklyn, Queens가 대다수)
"""

display(train_df["neighbourhood_group"].value_counts())
sns.countplot(train_df, x="neighbourhood_group")
```

```python
"""
지역별 가격 정보 확인

- 공통적으로 가격이 $500 이하의 데이터가 대다수이다.
- Manhattan의 경우 가격이 $500 보다 더 비싼 숙소도 어느정도 있다. ($500~$1000)
"""

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
x_limit = [0, 3000]
# price가 3000 보다 높은 데이터는 38개 인데, 38개의 데이터는 안봐도 분포를 살펴보는데 문제가 없어서 날린다.
# 확인: len(train_df[train_df["price"] > 3000])

ax1 = sns.histplot(train_df[train_df["neighbourhood_group"] == "Manhattan"], x="price", bins=200, ax=axes[0,0])
ax1.set_title("Manhattan price distribution")
ax1.set_xlim(x_limit)
    
ax2 = sns.histplot(train_df[train_df["neighbourhood_group"] == "Brooklyn"], x="price", bins=200, ax=axes[0,1])
ax2.set_title("Brooklyn price distribution")
ax2.set_xlim(x_limit)

ax3 = sns.histplot(train_df[train_df["neighbourhood_group"] == "Queens"], x="price", bins=200, ax=axes[0,2])
ax3.set_title("Queens price distribution")
ax3.set_xlim(x_limit)

ax4 = sns.histplot(train_df[train_df["neighbourhood_group"] == "Bronx"], x="price", bins=200, ax=axes[1,0])
ax4.set_title("Bronx price distribution")
ax4.set_xlim(x_limit)

ax5 = sns.histplot(train_df[train_df["neighbourhood_group"] == "Staten Island"], x="price", bins=200, ax=axes[1,1])
ax5.set_title("Staten Island price distribution")
ax5.set_xlim(x_limit)

fig.delaxes(axes[1][2])
```

#### last_review
- latest review

```python
"""
데이터 타입 변경 (object -> float)
"""

def yyyymmdd_to_num(yyyymmdd):
    if type(yyyymmdd) is str:
        return int(yyyymmdd.replace("-", ""))
    # nan은 변환하지 않는다
    return yyyymmdd

train_df["last_review_num"] = train_df["last_review"].apply(yyyymmdd_to_num)
train_df["last_review_num"]
```

```python
"""
최근 리뷰 날짜 데이터 분포 확인

- 2019년 데이터 자체가 많다
"""

sns.histplot(train_df, x="last_review_num", bins=200)
```

```python
"""
- 개인적인 직관에 따르면 가장 최근 리뷰 날짜가 가격과 직결되진 않을거라 생각된다.
- 데이터 분포도 날짜와 가격의 상관관계 패턴을 보기가 조금 힘든 것 같다.
  - 2019년 데이터 자체가 많기도 하다.

-> 일단은 피쳐로 사용하지 않기로 한다. (필요 시 이 데이터도 피쳐로 사용해보도록 한다)
"""

sns.scatterplot(train_df, x="last_review_num", y="price")
```
