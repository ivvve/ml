```python
data = load_wine()

dir(data)
```

## (3) ë°ì´í„° ì´í•´í•˜ê¸°
ì§€í”¼ì§€ê¸°ë©´ ë°±ì „ë¶ˆíƒœ! ë‹¤ë£¨ì–´ì•¼ í•  ë°ì´í„°ë¥¼ ìì„¸íˆ ì‚´í´ë´…ì‹œë‹¤.

```python
import pandas as pd

feature_data = data.data
label_data = data.target

feature_data.shape, label_data.shape
```

## (4) train, test ë°ì´í„° ë¶„ë¦¬
ëª¨ë¸ í•™ìŠµê³¼ í…ŒìŠ¤íŠ¸ìš© ë¬¸ì œì§€ì™€ ì •ë‹µì§€ë¥¼ ì¤€ë¹„í•´ ë´…ì‹œë‹¤.  
X_train, X_test, y_train, y_testë¥¼ ìƒì„±í•˜ëŠ” ë°©ë²•ì„ ì°¸ê³ í•´ ë³´ì„¸ìš”.


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(feature_data, label_data,
                                                    test_size=0.2, stratify=label_data,
                                                    random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## (5) ë‹¤ì–‘í•œ ëª¨ë¸ë¡œ í•™ìŠµì‹œì¼œë³´ê¸°
í•™ìŠµë°ì´í„° X_train, y_train ì„ í™œìš©í•´ ë¶„ë¥˜ê¸° ëª¨ë¸ì„ ë§Œë“¤ì–´ ë´…ì‹œë‹¤. ì–´ë–¤ ëª¨ë¸ì´ ê°€ì¥ ì¢‹ì€ ì„±ëŠ¥ì„ ë³´ì¼ê¹Œìš”?

- Decision Tree ì‚¬ìš©í•´ ë³´ê¸°
- Random Forest ì‚¬ìš©í•´ ë³´ê¸°
- SVM ì‚¬ìš©í•´ ë³´ê¸°
- SGD Classifier ì‚¬ìš©í•´ ë³´ê¸°
- Logistic Regression ì‚¬ìš©í•´ ë³´ê¸°

```python
from sklearn.tree import DecisionTreeClassifier

# í•™ìŠµ
DecisionTreeClassifier_model = DecisionTreeClassifier()
DecisionTreeClassifier_model.fit(X_train, y_train)

# í‰ê°€
y_pred = DecisionTreeClassifier_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

# í•™ìŠµ
RandomForestClassifier_model = RandomForestClassifier()
RandomForestClassifier_model.fit(X_train, y_train)

# í‰ê°€
y_pred = RandomForestClassifier_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC

# í•™ìŠµ
SVC_model = SVC(probability=True)
SVC_model.fit(X_train, y_train)

# í‰ê°€
y_pred = SVC_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import SGDClassifier

# í•™ìŠµ
SGDClassifier_model = SGDClassifier()
SGDClassifier_model.fit(X_train, y_train)

# í‰ê°€
y_pred = SGDClassifier_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

# í•™ìŠµ
LogisticRegression_model = LogisticRegression()
LogisticRegression_model.fit(X_train, y_train)

# í‰ê°€
y_pred = LogisticRegression_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

## (6) ëª¨ë¸ì„ í‰ê°€í•´ ë³´ê¸°
í•™ìŠµëœ ëª¨ë¸ë“¤ì˜ í…ŒìŠ¤íŠ¸ë°ì´í„° ì˜ˆì¸¡ ê²°ê³¼ë¥¼ ì–´ë–»ê²Œ í•´ì„í•´ì•¼ í• ê¹Œìš”? ëª¨ë¸ì˜ ì„±ëŠ¥ì„ í‰ê°€í•˜ëŠ” ì§€í‘œë¡œëŠ” ë¬´ì—‡ì´ ì¢‹ì„ê¹Œìš”?  
sklearn.metrics ì—ì„œ ì œê³µí•˜ëŠ” í‰ê°€ì§€í‘œ ì¤‘ ì ì ˆí•œ ê²ƒì„ ì„ íƒí•´ ë³´ì„¸ìš”. ì„ íƒí•˜ì‹  ì´ìœ ë„ ì„¤ëª…í•´ ì£¼ì„¸ìš”.

- ì™€ì¸ ë“±ê¸‰ì„ ì˜ˆì¸¡í•˜ëŠ” ëª¨ë¸ì´ê¸° ë•Œë¬¸ì— ë˜ë„ë¡ì´ë©´ ì •í™•í•œ ë“±ê¸‰ì´ í•„ìš”í•  ê²ƒì´ë‹¤.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

y_pred = RandomForestClassifier_model.predict(X_test)
y_pred_proba = RandomForestClassifier_model.predict_proba(X_test)

print(f"Accuracy Score: {accuracy_score(y_test, y_pred)}")
print(f"Precision Score: {precision_score(y_test, y_pred, average='weighted')}")
print(f"Recall Score: {recall_score(y_test, y_pred, average='weighted')}")
print(f"F1 Score: {f1_score(y_test, y_pred, average='weighted')}")
print(f"ROC AUC Score: {roc_auc_score(y_test, y_pred_proba, multi_class='ovo')}")
```
